```python
import functionlib as fl
import numpy as np
import matplotlib.pyplot as plt
```

```python
spot = np.linspace(50,150,1000)

T_list = [0.5, 1, 1.5, 2]

K = 100
r = 0.035
sigma = 0.2

premium_list = [[]]*4

for i in range(len(T_list)):
    premium_list[i] = [fl.black_scholes(S, K, T_list[i], r, sigma) for S in spot]

    plt.plot(spot, premium_list[i], label=f"Maturit = {T_list[i]}")

plt.xlabel("Spot")
plt.ylabel("Premium")
plt.title("Premium = f(Spot) with different maturities")
plt.grid()
plt.legend()
plt.show()


```
