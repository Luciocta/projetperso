```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pdi_payoff(S, K, B):
    if S < B:
        return max(K-S,0)
    else:
        return 0
S_values = np.linspace(0,200,1000)
pdi_payoff_values = [pdi_payoff(S, 70, 50) for S in S_values]
plt.plot(S_values, pdi_payoff_values)
plt.show()
```

```python
put = lambda S, K: max(K-S, 0)

def spread(K1, K2, S):
    if S <= K1:
        return K2-K1
    if S > K1 and S <= K2:
        return K2 - S
    else:
        return 0

spread_payoff = [spread(45, 50, S)*4 for S in S_values]

replication_payoff = [spread(48, 50, S)*10 + put(S, 50) for S in S_values]
plt.plot(S_values, replication_payoff, color='r', marker='.')
plt.plot(S_values, pdi_payoff_values)
plt.show()
```
