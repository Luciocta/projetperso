```python
import matplotlib.pyplot as plt
import functionlib as fl
import numpy as np
```

```python
S0 = 100
T = 1
r = 0.03
vol = 0.2

gammastrike = lambda strike, vol: fl.gamma(S0, strike, T, r, vol)

strike_range = np.linspace(50,150,100)
L_gamma = np.zeros((5,100))
volatility_range = [0.1, 0.15, 0.2, 0.25, 0.3]

for i, vol in enumerate(volatility_range):
    L_gamma[i] = [gammastrike(strike, vol) for strike in strike_range]
    plt.plot(strike_range, L_gamma[i], label = f'Volatlity = {vol}')

plt.legend()
plt.grid()
plt.show()
```

```python
S0 = 100
#T = 1
r = 0.03
sigma = 0.2

maturity_range = range(1,5)
strike_range = np.linspace(50,150,100)
L_gamma = np.zeros((5,100))

for i, maturity in enumerate(maturity_range):
    L_gamma[i] = [fl.gamma(S0, strike, maturity, r , sigma) for strike in strike_range]
    plt.plot(strike_range, L_gamma[i], label = f'Maturity = {maturity}')

plt.legend()
plt.grid()
plt.show()
```
