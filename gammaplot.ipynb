```python
import matplotlib.pyplot as plt
import functionlib as fl
import numpy as np
```

```python
S0 = 100
T = 1
r = 0.03
vol = 0.2

gammastrike = lambda strike: fl.gamma(S0, strike, T, r, vol)

strike_range = np.linspace(50,150,100)
L_gamma = [gammastrike(strike) for strike in strike_range]
plt.plot(strike_range, L_gamma)
plt.grid()
plt.show()
```
